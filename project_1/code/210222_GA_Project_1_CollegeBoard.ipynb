#### 6. Fix Incorrect Data Types
Based on what you discovered above, use appropriate methods to re-type incorrectly typed data.
- Define a function that will allow you to convert participation rates to an appropriate numeric type. Use `map` or `apply` to change these columns in each dataframe.

````python
def to_float(x):
    '''
    Function changes var x into a float, but to skip over NaNs if the var is a NaN instead of introducing an error.
    Returns modified float var.
    '''
    try:
        out = float(x)
    except:
        out = np.nan
    return out

def change_participation_into_a_float(participation):
    '''
    Function takes a Series and takes 'xx%'-format strings and replaces the series with equivalent 'x.xx' 
    Requires to_float() to be used.
    Returns modified Series.
    '''
    try:
        participation = participation.apply(lambda x: x.strip('%'))
    except:
        participation = participation.apply(lambda x: x)
    participation = participation.apply(lambda x: (to_float(x) / 100))
    return participation
````

````python
# participation is a % and should be changed to a float between 0 and 100 to facilitate %-based arithmetic later on.

sat_2017['Participation'] = change_participation_into_a_float(sat_2017['Participation'])
act_2017['Participation'] = change_participation_into_a_float(act_2017['Participation'])
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def calc_std_dev(series):
    '''
    Function takes a pandas.core.series.Series object and finds the standard deviation of the data set.
    Returns standard deviation according to the given formula.
    '''
    summation = 0
    for item in series:
        summation += ((item - series.mean())**2)
    return (summation / (series.count())) ** .5
````

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    '''
    Function creates a multi-plot histogram.
    dataframe: Dataframe to use.
    list_of_columns: List of names of dataframe's series to use in each subplot
    list_of_titles: List of title of each series in list_of_columns. To be displayed above subplot.
    list_of_xlabels: List of x-axis titles of each series in list_of_columns. To be displayed below each subplot's x-axis.
    list_of_legend_labels (Optional): List of each series name within list_of_columns. For when a
        subplot contains more than 1 series. 
    Does not return a value.
    '''
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(12,12)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column]) # feel free to add more settings
        # Set titles, labels, etc here for each subplot
        ax[i].set_xlabel(list_of_xlabels[i])
        ax[i].set_ylabel('No. of states')
        ax[i].set_title(list_of_titles[i])
````
